# Send feedback to LangSmith

```python
# In streamlit app the main challange is to get the right run_id, it has to be exposed through fastapi backend
```

```python
from langsmith import Client

client = Client()
client.create_feedback(
    run_id="8a3e70fb-efba-4e9f-a9ed-49c408d3f645", # here run_id is from the LangSmith trace (parent run_id, which is basically the same as the trace_id), needs to be extracted in the workflow
    key="user_feedback",
    score=0,
    feedback_source_type="api",
)
```

```python
client.create_feedback(
    run_id="8a3e70fb-efba-4e9f-a9ed-49c408d3f645", 
    key="user_feedback_1",
    score=1,
    feedback_source_type="api",
)
```
